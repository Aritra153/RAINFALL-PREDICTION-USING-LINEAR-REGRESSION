```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
dataset2=pd.read_csv("rainfall_final.csv")
dataset2
```

```python
dataset2.drop(["Unnamed: 0"],axis=1)
#As we need to predict the value "PrecipitationSumInches",so we separate the column out of the dataset and store it in a Variable Y
X=dataset2.drop(["Unnamed: 0"],axis=1).drop(["PrecipitationSumInches"],axis=1)
Y=dataset2.drop(["Unnamed: 0"],axis=1)["PrecipitationSumInches"]
```

```python
#reshaping Y into a vector
Y=Y.values.reshape(-1,1)
Y
```

```python
day_index=805
days=[i for i in range(Y.size)]
```

```python
#initializing the linear regression classifier
clf=LinearRegression()

#training the classifier
clf.fit(X,Y)
```

```python
#plotting precipitation vs days(with 805th day marked separately)
print("Precipitation vs Days")
plt.scatter(days,Y,color='g')
plt.scatter(days[day_index],Y[day_index],color='r')
plt.xlabel("Days")
plt.ylabel("Precipitation")
plt.show()
```

```python
#predicting the amount of rainfall(precipitationSumInches) for a ceratin value of the inputs
inp = np.array([[76], [49], [35], [34.0], [35.0], [28.0], [76.0], [59.0],
                [33], [33.56], [32.49], [10], [10], [10], [8], [3], [14]])
inp=inp.reshape(1,-1)
print("The Precipitation in Inches for the set of input is:")
print(-clf.predict(inp))
```
